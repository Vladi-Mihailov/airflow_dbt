```python
import pandas as pd
from clickhouse_driver import Client
from datetime import date
from sqlalchemy import create_engine, text
from urllib.parse import quote_plus
import psycopg2
from psycopg2.extras import execute_values
import logging
import time
from tqdm import tqdm

# =========================================================
# LOGGING
# =========================================================
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)-8s | %(message)s",
    datefmt="%H:%M:%S"
)
log = logging.getLogger(__name__)

# =========================================================
# CONFIG
# =========================================================
CSV_PATH = r"C:\Users\Vladimir\Desktop\work\test_task\Production\Data task new P.txt"

# ClickHouse
CH_CONN = {
    "host": "localhost",
    "database": "default"
}

# PostgreSQL
PG_USER = "postgres"
PG_PASSWORD = "admin"
PG_HOST = "localhost"
PG_PORT = "5432"
PG_DBNAME = "postgres"

PG_CONN_STR = (
    f"postgresql+psycopg2://"
    f"{quote_plus(PG_USER)}:{quote_plus(PG_PASSWORD)}"
    f"@{PG_HOST}:{PG_PORT}/{PG_DBNAME}"
)

# =========================================================
# START
# =========================================================
start_total = time.perf_counter()
log.info("Старт загрузки данных")

# =========================================================
# READ CSV
# =========================================================
log.info("Чтение CSV")
t0 = time.perf_counter()

df = pd.read_csv(
    CSV_PATH,
    sep="\t",
    engine="python",
    encoding="utf-8-sig",
    header=0
)

df.columns = [
    "client_id", "year", "month",
    "channel", "income", "volume", "trans_cnt"
]

df = df.apply(lambda c: c.str.strip() if c.dtype == "object" else c)
df["income"] = df["income"].str.replace(",", ".").astype(float)
df["volume"] = df["volume"].str.replace(",", ".").astype(float)
df["trans_cnt"] = df["trans_cnt"].astype(int)
df["period"] = df.apply(
    lambda r: date(int(r["year"]), int(r["month"]), 1), axis=1
)

log.info("CSV загружен: %d строк (%.2f сек)",
         len(df), time.perf_counter() - t0)

# =========================================================
# CLICKHOUSE
# =========================================================
log.info("ClickHouse: инициализация")
ch = Client(**CH_CONN)

ch.execute("""
CREATE TABLE IF NOT EXISTS test_tasks.fact_client_metrics
(
    client_id UInt64,
    period Date,
    channel LowCardinality(String),
    income Decimal(18,4),
    volume Decimal(18,4),
    trans_cnt UInt32
)
ENGINE = MergeTree
PARTITION BY toYYYYMM(period)
ORDER BY (period, channel, client_id);
""")

log.info("ClickHouse: проверка существующих данных")
t0 = time.perf_counter()

existing_keys_ch = ch.execute(
    "SELECT client_id, period, channel FROM test_tasks.fact_client_metrics"
)
existing_set_ch = set(existing_keys_ch)

df_new_ch = df[
    ~df.apply(
        lambda r: (r.client_id, r.period, r.channel) in existing_set_ch,
        axis=1
    )
]

log.info("ClickHouse: найдено %d новых строк (%.2f сек)",
         len(df_new_ch), time.perf_counter() - t0)

if not df_new_ch.empty:
    log.info("ClickHouse: загрузка данных")

    data_ch = list(
        tqdm(
            df_new_ch[
                ["client_id", "period", "channel",
                 "income", "volume", "trans_cnt"]
            ].itertuples(index=False, name=None),
            desc="ClickHouse insert",
            unit=" rows"
        )
    )

    ch.execute(
        "INSERT INTO test_tasks.fact_client_metrics VALUES",
        data_ch
    )

    log.info("ClickHouse: загружено %d строк", len(data_ch))
else:
    log.info("ClickHouse: новых данных нет")

# =========================================================
# POSTGRESQL
# =========================================================
log.info("PostgreSQL: инициализация")
engine = create_engine(PG_CONN_STR)

with engine.connect() as conn:
    conn.execute(text("""
    CREATE TABLE IF NOT EXISTS fact_client_metrics (
        client_id BIGINT NOT NULL,
        period DATE NOT NULL,
        channel TEXT NOT NULL,
        income NUMERIC(18,4),
        volume NUMERIC(18,4),
        trans_cnt INTEGER,
        PRIMARY KEY (client_id, period, channel)
    );
    """))
    conn.execute(
        text("CREATE INDEX IF NOT EXISTS idx_period ON fact_client_metrics(period);")
    )
    conn.execute(
        text("CREATE INDEX IF NOT EXISTS idx_channel ON fact_client_metrics(channel);")
    )

    log.info("PostgreSQL: проверка существующих данных")
    t0 = time.perf_counter()

    existing_keys_pg = conn.execute(
        text("SELECT client_id, period, channel FROM fact_client_metrics")
    ).fetchall()

existing_set_pg = set(existing_keys_pg)

df_new_pg = df[
    ~df.apply(
        lambda r: (r.client_id, r.period, r.channel) in existing_set_pg,
        axis=1
    )
]

log.info("PostgreSQL: найдено %d новых строк (%.2f сек)",
         len(df_new_pg), time.perf_counter() - t0)

if not df_new_pg.empty:
    log.info("PostgreSQL: загрузка данных")

    records = list(
        tqdm(
            df_new_pg[
                ["client_id", "period", "channel",
                 "income", "volume", "trans_cnt"]
            ].itertuples(index=False, name=None),
            desc="PostgreSQL insert",
            unit=" rows"
        )
    )

    conn_pg = psycopg2.connect(
        dbname=PG_DBNAME,
        user=PG_USER,
        password=PG_PASSWORD,
        host=PG_HOST,
        port=PG_PORT
    )
    cur_pg = conn_pg.cursor()

    execute_values(
        cur_pg,
        """
        INSERT INTO fact_client_metrics
        (client_id, period, channel, income, volume, trans_cnt)
        VALUES %s
        ON CONFLICT (client_id, period, channel) DO NOTHING
        """,
        records,
        page_size=10_000
    )

    conn_pg.commit()
    cur_pg.close()
    conn_pg.close()

    log.info("PostgreSQL: загружено %d строк", len(records))
else:
    log.info("PostgreSQL: новых данных нет")

# =========================================================
# FINISH
# =========================================================
log.info("Завершено за %.2f сек",
         time.perf_counter() - start_total)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from clickhouse_driver import Client
from datetime import date

# ====================================================
#  Подключение к ClickHouse и загрузка данных
# ====================================================
CH_CONN = {
    "host": "localhost",
    "database": "default"
}

ch = Client(**CH_CONN)

query = """
SELECT
    client_id,
    period,
    channel,
    income,
    volume,
    trans_cnt
FROM test_tasks.fact_client_metrics
"""

data = ch.execute(query)

df = pd.DataFrame(
    data,
    columns=['client_id', 'period', 'channel', 'income', 'volume', 'trans_cnt']
)

df['period'] = pd.to_datetime(df['period'])

# ====================================================
# Уникальные клиенты по периодам
# ====================================================
unique_clients = (
    df.groupby('period')['client_id']
      .nunique()
      .reset_index(name='unique_clients')
)

# ====================================================
#  Churn / прирост клиентов
# ====================================================
clients_by_period = (
    df.groupby('period')['client_id']
      .apply(set)
      .reset_index(name='clients')
)

churn_data = []

for i in range(1, len(clients_by_period)):
    prev_clients = clients_by_period.loc[i - 1, 'clients']
    curr_clients = clients_by_period.loc[i, 'clients']

    churn_data.append({
        'curr_period': clients_by_period.loc[i, 'period'],
        'new_clients': len(curr_clients - prev_clients),
        'lost_clients': len(prev_clients - curr_clients),
        'net_growth': len(curr_clients) - len(prev_clients)
    })

churn_df = pd.DataFrame(churn_data)

# ====================================================
#  Активность клиентов по периодам
# ====================================================
activity = (
    df.groupby('period')
      .agg(
          trans_cnt=('trans_cnt', 'sum'),
          volume=('volume', 'sum'),
          income=('income', 'sum')
      )
      .reset_index()
)

# ====================================================
# ВИЗУАЛИЗАЦИИ
# ====================================================

# --- График 1: уникальные клиенты по периодам с min/max ---
plt.figure(figsize=(10,5))
line_color = 'tab:blue'

plt.plot(
    unique_clients['period'],
    unique_clients['unique_clients'],
    marker='o',
    linewidth=2,
    color=line_color,
    alpha=0.65
)

plt.title('Уникальные клиенты по периодам')
plt.xlabel('Период')
plt.ylabel('Количество клиентов')
plt.grid(True)

y_min = unique_clients['unique_clients'].min() * 0.95
y_max = unique_clients['unique_clients'].max() * 1.05
plt.ylim(y_min, y_max)

min_idx = unique_clients['unique_clients'].idxmin()
max_idx = unique_clients['unique_clients'].idxmax()

min_val = unique_clients.loc[min_idx, 'unique_clients']
max_val = unique_clients.loc[max_idx, 'unique_clients']
min_period = unique_clients.loc[min_idx, 'period']
max_period = unique_clients.loc[max_idx, 'period']

plt.scatter([min_period, max_period], [min_val, max_val],
            color=line_color, zorder=5)

plt.text(min_period, min_val,
         f'{min_val:,.0f}'.replace(',', ' '),
         ha='right', va='bottom', fontsize=10)

plt.text(max_period, max_val,
         f'{max_val:,.0f}'.replace(',', ' '),
         ha='left', va='bottom', fontsize=10)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- График 2: чистый прирост клиентов (обновлённый) ---
plt.figure(figsize=(10,5))

bar_width = pd.Timedelta(days=20)

plt.bar(
    churn_df['curr_period'],
    churn_df['net_growth'],
    color='skyblue',
    width=bar_width,
    alpha=0.65
)

plt.title('Чистый прирост клиентов по периодам')
plt.xlabel('Период')
plt.ylabel('Прирост')
plt.grid(True, axis='y')

min_idx = churn_df['net_growth'].idxmin()
max_idx = churn_df['net_growth'].idxmax()

min_val = churn_df.loc[min_idx, 'net_growth']
max_val = churn_df.loc[max_idx, 'net_growth']
min_period = churn_df.loc[min_idx, 'curr_period']
max_period = churn_df.loc[max_idx, 'curr_period']

plt.text(min_period, min_val,
         f'{min_val:,.0f}'.replace(',', ' '),
         ha='center', va='bottom', fontsize=10)

plt.text(max_period, max_val,
         f'{max_val:,.0f}'.replace(',', ' '),
         ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- График 3: активность клиентов по периодам (масштабированные показатели) ---
plt.figure(figsize=(10,5))

trans_scaled = activity['trans_cnt'] / 1_000
volume_scaled = activity['volume'] / 1_000000
income_scaled = activity['income'] / 1_000000

plt.plot(activity['period'], trans_scaled,
         marker='o', linewidth=2, label='Транзакции (тыс.)', alpha=0.65)

plt.plot(activity['period'], volume_scaled,
         marker='o', linewidth=2, label='Объём (млн.)', alpha=0.65)

plt.plot(activity['period'], income_scaled,
         marker='o', linewidth=2, label='Доход (млн.)', alpha=0.65)

plt.title('Активность клиентов по периодам')
plt.xlabel('Период')
plt.ylabel('')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



# ====================================================
#  МИГРАЦИЯ КЛИЕНТОВ МЕЖДУ КАНАЛАМИ
# ====================================================






# Функция для переноса слов и заглавных букв
def wrap_and_capitalize(labels):
    return ['\n'.join([word.capitalize() for word in label.split(' ')]) for label in labels]

plt.figure(figsize=(10, 7))

ax = sns.heatmap(
    migration_share * 100,          # умножаем на 100 для легенды
    annot=True,
    fmt=".1f",                      # просто число
    cmap="Blues",
    xticklabels=wrap_and_capitalize(migration_share.columns),
    yticklabels=wrap_and_capitalize(migration_share.index),
    cbar_kws={'format': '%.0f%%'},  # легенда в процентах
)

# Добавляем % к аннотациям вручную
for text in ax.texts:
    text.set_text(text.get_text() + '%')

# Центрирование подписей X
plt.setp(ax.get_xticklabels(), ha='center')

# Подписи Y: горизонтальные, выравнивание по правому краю
plt.setp(ax.get_yticklabels(), rotation=0, ha='right')

plt.title('Миграция клиентов между каналами')
plt.xlabel('Текущий канал')
plt.ylabel('Предыдущий канал')
plt.tight_layout()
plt.show()












# ====================================================
# Считаем средний доход на клиента по каналам и периодам
# ====================================================
channel_client_income = (
    df.groupby(['period', 'channel'])
      .agg(
          total_income=('income', 'sum'),
          unique_clients=('client_id', 'nunique')
      )
      .reset_index()
)

channel_client_income['avg_income_per_client'] = channel_client_income['total_income'] / channel_client_income['unique_clients']

# ====================================================
#  Визуализация динамики среднего дохода на клиента
# ====================================================
plt.figure(figsize=(12,6))

channels = channel_client_income['channel'].unique()
palette = sns.color_palette("tab10", n_colors=len(channels))  # палитра для линий

for ch, color in zip(channels, palette):
    data_ch = channel_client_income[channel_client_income['channel'] == ch].sort_values('period')
    
    # Основная линия с прозрачностью 65%
    plt.plot(
        data_ch['period'], 
        data_ch['avg_income_per_client'] / 1_000,  # отображаем в тыс.
        marker='o',
        linewidth=2,
        alpha=0.65,
        color=color,
        label=ch.title()  # <-- Название канала с большой буквы
    )
    
    # Минимум
    min_idx = data_ch['avg_income_per_client'].idxmin()
    min_row = data_ch.loc[min_idx]
    plt.scatter(min_row['period'], min_row['avg_income_per_client']/1_000, color=color, zorder=5)
    plt.annotate(
        f"{min_row['avg_income_per_client'] / 1_000:.1f}",
        (min_row['period'], min_row['avg_income_per_client']/1_000),
        xytext=(0, -15),
        textcoords='offset points',
        ha='center',
        color='black'
    )
    
    # Максимум
    max_idx = data_ch['avg_income_per_client'].idxmax()
    max_row = data_ch.loc[max_idx]
    plt.scatter(max_row['period'], max_row['avg_income_per_client']/1_000, color=color, zorder=5)
    plt.annotate(
        f"{max_row['avg_income_per_client'] / 1_000:.1f}",
        (max_row['period'], max_row['avg_income_per_client']/1_000),
        xytext=(0, 10),
        textcoords='offset points',
        ha='center',
        color='black'
    )

plt.title('Динамика среднего дохода на клиента по каналам (тыс.)')
plt.xlabel('Период')
plt.ylabel('Средний доход на клиента, тыс.')
plt.legend(title='Канал')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()









# ====================================================
#  Уникальные клиенты по каналу и периоду
# ====================================================
clients_per_channel_period = (
    df.groupby(['period', 'channel'])['client_id']
      .nunique()
      .reset_index(name='unique_clients')
)

# ====================================================
#  Визуализация динамики уникальных клиентов с min/max
# ====================================================
plt.figure(figsize=(12,6))

channels = clients_per_channel_period['channel'].unique()

# Создаем цикл цветов
colors = plt.cm.tab10.colors  # 10 базовых цветов
color_dict = {ch: colors[i % len(colors)] for i, ch in enumerate(channels)}

for ch in channels:
    data_ch = clients_per_channel_period[clients_per_channel_period['channel'] == ch]
    color = color_dict[ch]

    # Основная линия с прозрачностью 65%
    plt.plot(
        data_ch['period'],
        data_ch['unique_clients'],
        marker='o',
        linewidth=2,
        label=ch,
        alpha=0.65,
        color=color
    )

    # Найдем min и max
    min_idx = data_ch['unique_clients'].idxmin()
    max_idx = data_ch['unique_clients'].idxmax()

    min_val = data_ch.loc[min_idx, 'unique_clients']
    max_val = data_ch.loc[max_idx, 'unique_clients']

    min_period = data_ch.loc[min_idx, 'period']
    max_period = data_ch.loc[max_idx, 'period']

    # Подсветка точек с прозрачностью 65%
    plt.scatter(min_period, min_val, color=color, alpha=0.65, zorder=5)
    plt.scatter(max_period, max_val, color=color, alpha=0.65, zorder=5)

    # Подписи чисел черным цветом
    plt.text(min_period, min_val, f'{min_val:,}'.replace(',', ' '), ha='center', va='bottom', color='black', fontsize=10)
    plt.text(max_period, max_val, f'{max_val:,}'.replace(',', ' '), ha='center', va='bottom', color='black', fontsize=10)

plt.title('Динамика уникальных клиентов по каналам')
plt.xlabel('Период')
plt.ylabel('Количество уникальных клиентов')
plt.legend(title='Канал')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

















# --- 3. Первый период для каждого клиента ---
first_period = (
    df.groupby('client_id')['period']
    .min()
    .reset_index(name='first_period')
)

# --- 4. Канал в первый период клиента ---
first_period = first_period.merge(
    df[['client_id', 'period', 'channel']],
    left_on=['client_id', 'first_period'],
    right_on=['client_id', 'period'],
    how='left'
)

# --- 5. Фильтр по каналу "Индивидуальные условия" ---
first_period_ind = first_period[
    first_period['channel'] == 'Индивидуальные условия'
]

# --- 6. Считаем новых клиентов по периодам ---
new_clients_dyn = (
    first_period_ind
    .groupby('first_period')['client_id']
    .nunique()
    .reset_index(name='new_clients')
)

# --- 7. Все периоды, которые реально есть в данных ---
all_periods = (
    df['period']
    .drop_duplicates()
    .sort_values()
    .to_frame(name='period')
)

# --- 8. Добавляем периоды без новых клиентов ---
new_clients_dyn = (
    all_periods
    .merge(
        new_clients_dyn,
        left_on='period',
        right_on='first_period',
        how='left'
    )
    .drop(columns='first_period')
    .fillna({'new_clients': 0})
)

# --- 9. Минимум и максимум ---
min_idx = new_clients_dyn['new_clients'].idxmin()
max_idx = new_clients_dyn['new_clients'].idxmax()

min_row = new_clients_dyn.loc[min_idx]
max_row = new_clients_dyn.loc[max_idx]

# --- 10. График ---
plt.figure(figsize=(12, 6))

line_color = 'tab:blue'

plt.plot(
    new_clients_dyn['period'],
    new_clients_dyn['new_clients'],
    marker='o',
    linewidth=2,
    color=line_color,
    alpha=0.65
)

# Диапазон по оси Y для корректного смещения подписей
y_min, y_max = new_clients_dyn['new_clients'].min(), new_clients_dyn['new_clients'].max()
y_range = y_max - y_min

# Минимум
dy_min = -15 if (min_row['new_clients'] - y_min) > 0.05 * y_range else 10
plt.scatter(min_row['period'], min_row['new_clients'], color=line_color, zorder=5)
plt.annotate(
    f"{int(min_row['new_clients'])}",
    (min_row['period'], min_row['new_clients']),
    xytext=(0, dy_min),
    textcoords='offset points',
    ha='center',
    color='black'
)

# Максимум
dy_max = 10 if (y_max - max_row['new_clients']) > 0.05 * y_range else -10
plt.scatter(max_row['period'], max_row['new_clients'], color=line_color, zorder=5)
plt.annotate(
    f"{int(max_row['new_clients'])}",
    (max_row['period'], max_row['new_clients']),
    xytext=(0, dy_max),
    textcoords='offset points',
    ha='center',
    color='black'
)

plt.title('Динамика новых клиентов по каналу "Индивидуальные условия"')
plt.xlabel('Период')
plt.ylabel('Количество новых клиентов')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ====================================================
# CLASSIC RETENTION (STRICT, NO REACTIVATION)
# ====================================================

# --- Когорта = первый период клиента в банке ---
df['cohort'] = df.groupby('client_id')['period'].transform('min')
df['cohort_str'] = df['cohort'].dt.strftime('%Y-%m')

# --- Месяц жизни клиента ---
df['month_number'] = (
    (df['period'].dt.year - df['cohort'].dt.year) * 12 +
    (df['period'].dt.month - df['cohort'].dt.month) + 1
)

last_period = df['period'].max()
all_cohorts = sorted(df['cohort_str'].unique())

# ====================================================
# ФУНКЦИЯ ДЛЯ ПОСТРОЕНИЯ RETENTION
# ====================================================
def plot_retention(activity_df, title):
    # Клиенты в M1 (основание когорты)
    cohort_clients = (
        activity_df[activity_df['month_number'] == 1]
        .groupby('cohort_str')['client_id']
        .apply(set)
        .to_dict()
    )

    retention_rows = []

    for cohort, clients_m1 in cohort_clients.items():
        max_month = (
            (last_period.year - pd.to_datetime(cohort + '-01').year) * 12 +
            (last_period.month - pd.to_datetime(cohort + '-01').month) + 1
        )

        active_clients = clients_m1.copy()

        for month in range(1, max_month + 1):
            current_clients = set(
                activity_df[
                    (activity_df['cohort_str'] == cohort) &
                    (activity_df['month_number'] == month)
                ]['client_id']
            )

            # STRICT retention — без возврата
            active_clients &= current_clients

            retention_rows.append({
                'cohort': cohort,
                'month_number': month,
                'retention': len(active_clients) / len(clients_m1)
            })

    retention_df = pd.DataFrame(retention_rows)

    retention_pivot = (
        retention_df
        .pivot(index='cohort', columns='month_number', values='retention')
        .reindex(all_cohorts)
    )

    # В проценты
    retention_pivot_100 = retention_pivot * 100

    annotations = retention_pivot_100.applymap(
        lambda x: f"{x:.0f}%" if pd.notna(x) else ""
    )

    plt.figure(figsize=(12, 6))
    ax = sns.heatmap(
        retention_pivot_100,
        annot=annotations,
        fmt="",
        cmap="YlGnBu",
        vmin=0,
        vmax=100,
        linewidths=0.5,
        cbar_kws={
            'label': 'Retention (%)',
            'format': '%.0f%%'
        }
    )

    ax.xaxis.set_ticks_position('top')
    ax.xaxis.set_label_position('top')
    ax.set_xticklabels([f"M{int(x)}" for x in retention_pivot.columns])
    ax.set_yticklabels(retention_pivot.index, rotation=0)

    plt.title(title, pad=30)
    plt.xlabel('Месяц жизни клиента')
    plt.ylabel('Когорта')
    plt.tight_layout()
    plt.show()

# ====================================================
# 1️⃣ RETENTION ПО ВСЕМУ БАНКУ
# ====================================================
activity_all = (
    df
    .groupby(['cohort_str', 'client_id', 'month_number'])
    .size()
    .reset_index(name='cnt')
)

plot_retention(
    activity_all,
    'Классический Retention по когортам (весь банк, без реактивации)'
)

# ====================================================
# 2️⃣ RETENTION ПО КАНАЛАМ
# ====================================================
for channel in df['channel'].unique():
    activity_channel = (
        df[df['channel'] == channel]
        .groupby(['cohort_str', 'client_id', 'month_number'])
        .size()
        .reset_index(name='cnt')
    )

    plot_retention(
        activity_channel,
        f'Retention rate по когортам — канал "{channel.title()}"'
    )
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ====================================================
# CLASSIC RETENTION (STRICT, NO REACTIVATION)
# сначала ВСЕ клиенты, затем ПО КАНАЛАМ
# ====================================================

# --- Когорта = первый период клиента ---
df['cohort'] = df.groupby('client_id')['period'].transform('min')
df['cohort_str'] = df['cohort'].dt.strftime('%Y-%m')

# --- Месяц жизни клиента ---
df['month_number'] = (
    (df['period'].dt.year - df['cohort'].dt.year) * 12 +
    (df['period'].dt.month - df['cohort'].dt.month) + 1
)

last_period = df['period'].max()
all_cohorts = sorted(df['cohort_str'].unique())

# ====================================================
# 1️⃣ RETENTION БЕЗ КАНАЛОВ (ВСЕ КЛИЕНТЫ)
# ====================================================

activity_all = (
    df.groupby(['cohort_str', 'client_id', 'month_number'])
      .size()
      .reset_index(name='cnt')
)

cohort_clients_all = (
    activity_all[activity_all['month_number'] == 1]
    .groupby('cohort_str')['client_id']
    .apply(set)
    .to_dict()
)

retention_rows = []

for cohort, clients_m1 in cohort_clients_all.items():
    max_month = (
        (last_period.year - pd.to_datetime(cohort + '-01').year) * 12 +
        (last_period.month - pd.to_datetime(cohort + '-01').month) + 1
    )

    active_clients = clients_m1.copy()

    for month in range(1, max_month + 1):
        current_clients = set(
            activity_all[
                (activity_all['cohort_str'] == cohort) &
                (activity_all['month_number'] == month)
            ]['client_id']
        )

        active_clients &= current_clients  # strict retention

        retention_rows.append({
            'cohort': cohort,
            'month_number': month,
            'active_clients': len(active_clients)
        })

retention_df_all = pd.DataFrame(retention_rows)

retention_pivot_all = (
    retention_df_all
    .pivot(index='cohort', columns='month_number', values='active_clients')
    .reindex(all_cohorts)
)

annotations_all = retention_pivot_all.applymap(
    lambda x: f"{int(x)}" if pd.notna(x) else ""
)

plt.figure(figsize=(14.4, 6))  # +20% ширины

ax = sns.heatmap(
    retention_pivot_all,
    annot=annotations_all,
    fmt="",
    cmap="YlGnBu",
    linewidths=0.5,
    cbar_kws={'label': 'Количество активных клиентов'}
)

ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')
ax.set_xticklabels([f"M{int(x)}" for x in retention_pivot_all.columns])
ax.set_yticklabels(retention_pivot_all.index, rotation=0)

plt.title('Retention по когортам — все клиенты', pad=30)
plt.xlabel('Месяц жизни клиента')
plt.ylabel('Когорта')
plt.tight_layout()
plt.show()

# ====================================================
# 2️⃣ RETENTION ПО КАНАЛАМ
# ====================================================

channels = df['channel'].unique()

for channel in channels:
    df_ch = df[df['channel'] == channel]

    activity = (
        df_ch
        .groupby(['cohort_str', 'client_id', 'month_number'])
        .size()
        .reset_index(name='cnt')
    )

    cohort_clients = (
        activity[activity['month_number'] == 1]
        .groupby('cohort_str')['client_id']
        .apply(set)
        .to_dict()
    )

    retention_rows = []

    for cohort, clients_m1 in cohort_clients.items():
        max_month = (
            (last_period.year - pd.to_datetime(cohort + '-01').year) * 12 +
            (last_period.month - pd.to_datetime(cohort + '-01').month) + 1
        )

        active_clients = clients_m1.copy()

        for month in range(1, max_month + 1):
            current_clients = set(
                activity[
                    (activity['cohort_str'] == cohort) &
                    (activity['month_number'] == month)
                ]['client_id']
            )

            active_clients &= current_clients  # strict retention

            retention_rows.append({
                'cohort': cohort,
                'month_number': month,
                'active_clients': len(active_clients)
            })

    retention_df = pd.DataFrame(retention_rows)

    retention_pivot = (
        retention_df
        .pivot(index='cohort', columns='month_number', values='active_clients')
        .reindex(all_cohorts)
    )

    annotations = retention_pivot.applymap(
        lambda x: f"{int(x)}" if pd.notna(x) else ""
    )

    plt.figure(figsize=(14.4, 6))  # +20% ширины

    ax = sns.heatmap(
        retention_pivot,
        annot=annotations,
        fmt="",
        cmap="YlGnBu",
        linewidths=0.5,
        cbar_kws={'label': 'Количество активных клиентов'}
    )

    ax.xaxis.set_ticks_position('top')
    ax.xaxis.set_label_position('top')
    ax.set_xticklabels([f"M{int(x)}" for x in retention_pivot.columns])
    ax.set_yticklabels(retention_pivot.index, rotation=0)

    plt.title(f'Retention по когортам — "{channel.title()}"', pad=30)
    plt.xlabel('Месяц жизни клиента')
    plt.ylabel('Когорта')
    plt.tight_layout()
    plt.show()
```
